# Webscraping

```python
import requests
from bs4 import BeautifulSoup
import urllib.request as url_req

# import packages
```

```python
# url 
url = 'https://en.wikipedia.org/wiki/List_of_state_and_union_territory_capitals_in_India'
```

```python
#Query the website and return the html to the variable 'page'
page = url_req.urlopen(url)
```

```python
soup = BeautifulSoup(page)
```

```python
print(soup.prettify())
```

```python
# Return content between opening and closing tag including tag.
soup.title
```

```python
soup.title.string   # only getting the string part and not the tags
```

```python
# getting all the links on the page

soup.find_all('a')
```

```python
all_links = soup.find_all('a')
for link in all_links:
    print(link.get('href'))
```

```python
# find the tables: we want capitals, we should identify the right table

all_tables = soup.find_all('table')
all_tables
```

```python
# selecting specific class

right_table = soup.find('table',class_='wikitable sortable')
right_table
```

#### extracting the information into dataframe

```python
# Generate lists

B=[]
C=[]
D=[]
E=[]
F=[]
G=[]

for row in right_table.findAll('tr'):
    cells = row.findAll('td')
    print(cells)
    if len(cells) == 6:
        B.append(cells[0].find(text=True))
        C.append(cells[1].find(text=True))
        D.append(cells[2].find(text=True))
        E.append(cells[3].find(text=True))
        F.append(cells[4].find(text=True))
        G.append(cells[5].find(text=True))
     
```

```python
import pandas as pd
df = pd.DataFrame()
df['State/UT']=B
df['Administrative_Capital'] = C
df['Legislative_Capital'] = D
df['Judiciary_Capital'] = E
df['Year_Capital'] = F
df['Former_Capital'] = G
df
```

```python
# removing \n from all columns
for i in df.columns:
    df[i] = df[i].str.replace('\n','')
df
```
